```python
import pandas as pd
```

```python
df = pd.read_csv('./csv_file/국민여행조사.csv', encoding='utf-8')
```

```python
df.head()
```

```python
A12 = []
A13 = []
A14 = []
for i in range(len(df)):
    if df['A12'][i] == ' ':
        continue
    else:
        A12.append(float(df['A12'][i]))
    if df['A13'][i] == ' ':
        continue
    else:
        A13.append(float(df['A13'][i]))
    if df['A14'][i] == ' ':
        continue
    else:
        A14.append(float(df['A14'][i]))
```

```python
A12 = sum(A12)/len(A12)
A13 = sum(A13)/len(A13)
A14 = sum(A14)/len(A14)
```

```python
A14
```

```python
A12_2 = []
A13_2 = []
A14_2 = []
for i in range(len(df)):
    if df['A12'][i] == ' ':
        A12_2.append(A12)
    else:
        A12_2.append(float(df['A12'][i]))
    if df['A13'][i] == ' ':
        A13_2.append(A13)
    else:
        A13_2.append(float(df['A13'][i]))
    if df['A14'][i] == ' ':
        A14_2.append(A14)
    else:
        A14_2.append(float(df['A14'][i]))
```

```python
df['A12'] = A12_2
df['A13'] = A13_2
df['A14'] = A14_2
```

```python
df['A13']
```

```python
label =[]
for i in range(len(df)):
    if (df.A12[i] + df.A13[i] + df.A14[i])/3 >= 4:
        label.append(1)
    else: 
        label.append(0)
label
```

```python
df['label'] = label
```

```python
df.label
```

```python
df_label = df.filter(['ID','label', 'BSEX', 'BAGE'])
```

```python
df_label = df_label[(df_label['BAGE'] == 2) | (df_label['BAGE'] == 3)]
df_label = df_label.reset_index(drop=True)
```

```python
df_label.to_csv('./csv_file/라벨링.csv', encoding='utf-8', index=False)
```
