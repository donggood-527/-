```python
import pandas as pd
```

```python
df_2019 = pd.read_csv('./신한카드 데이터/df_2019.csv')
```

```python
df_2019
```

```python
상반기 = []
for i in range(len(df_2019)):
    if df_2019['이용년월'][i] == 201901:
        상반기.append('상반기')
    elif df_2019['이용년월'][i] == 201902:
        상반기.append('상반기')
    elif df_2019['이용년월'][i] == 201903:
        상반기.append('상반기')
    elif df_2019['이용년월'][i] == 201904:
        상반기.append('상반기')
    elif df_2019['이용년월'][i] == 201905:
        상반기.append('상반기')
    elif df_2019['이용년월'][i] == 201906:
        상반기.append('상반기')
    else:
        상반기.append('하반기')
```

```python
df_2019['분기'] = 상반기
```

```python
df_2019_상반기 = df_2019[df_2019['분기'] == '상반기']
```

```python
df_2019_상반기.to_csv('./신한카드 데이터/df_2019_상반기.csv', index=False)
```
