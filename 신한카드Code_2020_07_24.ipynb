```python
import pandas as pd
```

```python
df_2017 = pd.read_csv('./신한카드 데이터/KCTI_NATIVE_2017.txt', encoding='cp949', sep='|')
df_2018 = pd.read_csv('./신한카드 데이터/KCTI_NATIVE_2018.txt', encoding='cp949', sep='|')
df_2019 = pd.read_csv('./신한카드 데이터/KCTI_NATIVE_2019.txt', encoding='cp949', sep='|')
df_2020 = pd.read_csv('./신한카드 데이터/KCTI_NATIVE_2020.txt', encoding='cp949', sep='|')
df_for = pd.read_csv('./신한카드 데이터/KCTI_FOREIGNER.txt', encoding='cp949', sep='|')
df_all = pd.read_csv('./신한카드 데이터/KCTI_NATIVE_ALL.txt', encoding='cp949', sep='|')
```

```python
def df_modify(df):
    df.columns = ['거주지', '가맹점', '가맹점2', '업종1', '업종2', '성별', '연령대', '이용년월', '요일',
                 '이용시간대', '취급액', '이용건수']
    return df
```

```python
df_2017 = df_modify(df_2017)
df_2018 = df_modify(df_2018)
df_2019 = df_modify(df_2019)
df_2020 = df_modify(df_2020)
df_all = df_modify(df_all)
```

```python
df_2017.to_csv('./Code/신용카드_2017.csv')
df_2018.to_csv('./Code/신용카드_2018.csv')
df_2019.to_csv('./Code/신용카드_2019.csv')
df_for.to_csv('./Code/신용카드_for.csv')
df_all.to_csv('./Code/신용카드_all.csv')
```

```python
def train_value(df):
    df = df.groupby('업종2')['취급액'].sum()
    df = pd.DataFrame(df)
    df.reset_index(inplace=True)
    df = df.sort_values('취급액', ascending=False)
    df = df.reset_index(drop=True)
    return df
```

```python
train_2017 = train_value(df_2017)
train_2018 = train_value(df_2018)
train_2019 = train_value(df_2019)
train_2020 = train_value(df_2020)
train_all = train_value(df_all)
```

```python
train = pd.merge(train_2019, train_2020, on='업종2', how='left')
```
